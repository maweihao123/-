```python
file= open('Walden.txt', 'r')
words=[]                        #存放取出的单词
for line in file:
    for word in line.replace('\n','').split(' '):  #把line中的换行符变为空，再以空格为间隔取出单词
        words.append(word)      #单词放到words中

map = {}
for word in words:
    if  word in map.keys():
        map[word] = map[word] + 1
    else:
        map[word] = 1                                
del map['']
map1 = sorted(map.items(), key=lambda x:x [1], reverse=True)  #字典按value大小排序
print(map1)


file.close()
```
