```python
file= open('Walden.txt', 'r')
words=[]                        #
for line in file:
    for word in line.replace('\n','').split(' '):  #line
        words.append(word)      #words

map = {}
for word in words:
    if  word in map.keys():
        map[word] = map[word] + 1
    else:
        map[word] = 1                                
del map['']
map1 = sorted(map.items(), key=lambda x:x [1], reverse=True)  #value
print(map1)


file.close()
```
